## Array.of

There are cases where you need to create dynamically an array and you should use Array.of

```javascript
const yetMoreNumbers = Array.of(1, 2);
console.log(yetMoreNumbers);
```

_______
## Array.from

In this case, it is possible to split a single string in every character that is composed of. It returns an array. Converts iterable objects into real arrays.

```javascript
const itemsList = 'Hi Manu'
console.log(itemsList);

const arrLiItems = Array.from(itemsList);
console.log(arrLiItems);
```

## Add data to arrays

### To append data

```javascript
const letters = ['a', 'b', 'c'];
letters.push('d');
letters;
```

### To add at the beginning

```javascript
const names = ['Manuel', 'Alonso'];
names;
```

```javascript
names.unshift('Viñas');
names;
```

### To remove the last element

```javascript
letters;
```

```javascript
letters.pop();
letters;
```
